```python
class ObjectLeg:
    def __init__(self, color) -> None:
        self.color = color
    def get_color(self):
        return self.color

class HandlerCls:
    def legend_artist(self, legend, orig_handle, fontsize, handlebox):
        x0, y0 = handlebox.xdescent, handlebox.ydescent
        width, height = handlebox.width, handlebox.height
        rect= mpatches.Rectangle((x0, y0), width, height, 
                                 facecolor=orig_handle.get_color(), 
                                 transform=handlebox.get_transform())
        handlebox.add_artist(rect)
        return rect
        
fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(x, 'g-')
ax.legend([ObjectLeg(color='r')], ['Line'],
          handler_map={ObjectLeg: HandlerCls()}
          )
```

```python
x = np.random.uniform(0, 1, (10, 10))
fig = plt.figure()

def plot_line(ax):
    ax.plot([1, 2, 3])
def plot_colorbar(ax):
    return ax.pcolormesh(x, cmap='viridis')

gs = fig.add_gridspec(1, 2)
g1 = gs[0].subgridspec(2, 1)

for g in g1:
    ax = fig.add_subplot(g)
    plot_line(ax)
    
g2 = gs[1].subgridspec(2, 2)
axes = []
for i in range(2):
    for j in range(2):
        ax = fig.add_subplot(g2[i, j])
        im = plot_colorbar(ax)
        axes.append(ax)
        
fig.colorbar(im, ax=axes)
```

```python
def _forward(x):
    return np.sqrt(x)


def _inverse(x):
    return x**2

N = 100
X, Y = np.mgrid[0:3:complex(0, N), 0:2:complex(0, N)]
Z1 = (1 + np.sin(Y * 10.)) * X**2
fig, ax = plt.subplots()

norm = mcolors.FuncNorm((_forward, _inverse), vmin=0, vmax=20)
pcm = ax.pcolormesh(X, Y, Z1, norm=norm, cmap='PuBu_r', shading='auto')
ax.set_title('FuncNorm(x)')
fig.colorbar(pcm, shrink=0.6)
plt.show()
```
