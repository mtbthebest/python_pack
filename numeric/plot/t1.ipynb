```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.arange(-20, 20, 1)
```

```python
x
```

```python
y = x**2 - 2*x + 1
z = 4*x + 3
```

```python
y
```

```python
width, height = 10, 4
fig, ax = plt.subplots(1, 1, figsize=(width, height))
ax.plot(x, y, "r-")
```

```python
width, height = 10, 4
fig, ax = plt.subplots(1, 1, figsize=(width, height))
ax.plot(x, y, color='cyan', linestyle='dashed', linewidth=2, marker='*', markersize=20)
```

```python
width, height = 10, 10
fig, ax = plt.subplots(1, 1, figsize=(width, height))
ax.plot(x, y, color='#FF0FAF', label="x-y")
ax.plot(x, z, color="#3F002B", label="x-z", linestyle="--", marker='2')
ax.set_xlim(-40, 40)
ax.set_ylim(-5, 500)
ax.set(title="Squared")
ax.legend()
```

```python
width, height = 10, 10
fig, ax = plt.subplots(1, 1, figsize=(width, height))
ax.plot(x, y, color='#FF0FAF', label="x-y")
ax.plot(x, z, color="#3F002B", label="x-z", linestyle="--", marker='2')
ax.set_xlim(-40, 40)
ax.set_ylim(-5, 500)
ax.set_xlabel("X")
ax.set_ylabel("f(X)")
ax.set(title="Squared")
ax.legend()
```

```python
a = ["A", "B", "C"]
b = [1, 4, 10]
fig = plt.figure()

ax = plt.subplot(1, 3, 1)
ax.plot(a, b)
ax.set_title("Home")
plt.subplot(1, 3, 2)
plt.bar(a, b)
plt.subplot(133)
plt.scatter(a, b)
plt.suptitle("Multiple plots")
```

```python
lines = plt.plot(a, b)
plt.setp(lines, color="r", linewidth=10, label="x", marker='|', markersize=20,
         alpha=0.2, mec="blue", markeredgewidth=10)
plt.legend()
```
