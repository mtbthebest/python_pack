```python
from matplotlib.legend_handler import HandlerPatch


class HandlerRect(HandlerPatch):
    
    def create_artists(self, legend, origin_handle, xdescent, ydescent, width, height, fontsize, trans):
        
        xy = xdescent - 2, ydescent
        print(xdescent, ydescent, width, height)
        p = mpatches.Rectangle(xy=xy, width=width+1, height=height + 15, lw=0.1, facecolor='green', alpha=0.3)
        
        self.update_prop(p, origin_handle, legend)
        p.set_transform(trans)
        
        return [p]

fig, ax = plt.subplots()

xy, width, height = (0,0), 2, 1
patch = mpatches.Rectangle(xy=xy, width=width, height=height, lw=10, facecolor='green', alpha=0.6, ec='red')
c = ax.add_patch(patch)
ax.grid(False)
ax.set_xlim(-2, 4)
ax.set_ylim(-2, 2)
ax.set_xticks([])
ax.set_yticks([])

legend = plt.legend([patch], ["Rectangle"], handler_map={mpatches.Rectangle: HandlerRect()})
```

```python
a = np.arange(100).reshape((10, 10))
fig, ax = plt.subplots(figsize=(4, 4), constrained_layout=True)
norm = mcolors.Normalize(vmin=0, vmax=100.)
im = ax.pcolormesh(a, norm=norm, cmap='viridis')
fig.colorbar(im, ax=ax)
```
