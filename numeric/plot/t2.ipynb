```python
grid = gridspec.GridSpec(1, 2)
fig = plt.figure()

x = np.linspace(-np.pi /2, 5 * np.pi /2, 100)
y = np.sin(x)
z = 2*x + 5
ax1 = fig.add_subplot(grid[0])
ax2 = fig.add_subplot(grid[1])

line1, = ax1.plot(x, y)
line2, = ax2.plot(x, z, )

plt.setp(line1, color='red')
plt.setp(line2, label='line')
ax1.set_title('SIN(X)')
ax2.set_title('LINE(X)')

ax1.set_ylabel('Y values', fontweight='bold', alpha=0.75)
ax2.set_xlabel('X', color='green', rotation=105, loc='left')

fig.suptitle('Plotting functions')
fig.supxlabel('X sup')
```
