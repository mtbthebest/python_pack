```python
genders = ['female', 'male']
locations = ['from Africa', 'from Asia', 'from Europe', 'from US']
browsers = ['uses Chrome', 'uses Firefox', 'uses IE', 'uses Safari']
enc = OneHotEncoder(categories=[genders, locations, browsers])
# feature
X = [['male', 'from US', 'uses Safari'], ['female', 'from Europe', 'uses Firefox']]
enc.fit(X)
enc.transform([['female', 'from Asia', 'uses Chrome']]).toarray()
```

```python
X = np.array([['dog'] * 5 + ['cat'] * 2 + ['lion'] * 10], dtype=object).T
enc = OneHotEncoder(min_frequency=4, handle_unknown="infrequent_if_exist").fit(X)
enc.categories_
```

```python
enc.infrequent_categories_
```

```python
enc.transform(np.array([['dog'], ['cat'], ["mouse"]])).toarray()
```

```python
X = np.array([['dog'] * 5 + ['cat'] * 2 + ['lion'] * 10], dtype=object).T
enc = OrdinalEncoder(min_frequency=3).fit(X)
enc.categories_
```

```python
enc.infrequent_categories_
```

```python
enc.transform(np.array([['dog'], ['cat']]))
```

```python
X = np.array([[ -3., 5., 15 ],
               [  0., 6., 14 ],
               [  6., 3., 11 ]])
enc = KBinsDiscretizer(n_bins=[3, 2, 4], encode="ordinal")
```

```python
enc.fit(X)
```

```python
enc.bin_edges_
```

```python
enc.transform(X)
```
