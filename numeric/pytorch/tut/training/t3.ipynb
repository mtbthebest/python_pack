```python
from torch.utils.data import Sampler as TorchSampler, DataLoader, Dataset
import torch
```

```python
class TestDat(Dataset):
    
    def __init__(self) -> None:
        self.seq = list(range(100))
        
    def __getitem__(self, index):
        return self.seq[index]
    
    def __len__(self):
        return len(self.seq)
```

```python

class Sampler(TorchSampler):
    
    def __init__(self, dataset, batch_size, seeds=[0]):
        print("ok")
        self.dataset = dataset
        data_len = len(dataset)
        
        num_samples = data_len // batch_size
        
        num_samples *= batch_size
        print("number samples is ", num_samples)
        
        self.num_samples = num_samples
        
        indices = list(range(len(dataset)))
        
        indices = torch.tensor(indices, dtype=torch.int64)
        
        self.indices = indices
        self.seeds = seeds
        self.epoch = 0  
        self.n = len(self.dataset)  
    
    def init_rng(self):
        rng = torch.Generator()
        
        rng.manual_seed(self.seeds[self.epoch])
        
        return rng
        
    def set_epoch(self, epoch):
        self.epoch = epoch
    
    def __iter__(self):
        rng = self.init_rng()
        print(torch.randperm(len(self.indices), generator=rng))
        for k, i in enumerate(torch.randperm(len(self.indices), generator=rng)):
            if k == self.num_samples:
                print("Done")
                break
            print(i, k, self.n)
            yield self.indices[i]
        

    def __len__(self):
        return self.num_samples

batch_size = 6
dst = TestDat() 
sampler = Sampler(dst, batch_size=6,)
loader = DataLoader(dst, batch_size=6, sampler=sampler)
```
